```python
# Penggunaan mahotas.otsu

import mahotas
import mahotas.demos
import numpy as np
from pylab import imshow,gray,show
from os import path

photo = mahotas.demos.load ('luispedro',as_grey=True)
photo = photo.astype (np.uint8)
gray()
imshow(photo)
show()
```
